## **Циклы**

Позволяют выполнять блок кода несколько раз; есть переменная которая изменяется на каждой *итерации* цикла.

Есть два вида циклов:
- `for`:
``` python
for *любое имя переменной* in range(*положительное целое число*): 
    *код который будет повторяться*
    *код который будет повторяться* 
*код вне цикла*
```
- `while`:
``` python
while *условие*: 
    *код который будет повторяться пока верно условие*
    *код который будет повторяться пока верно условие*
*код вне цикла*
```
Примеры использования, цикл `for`:

````python
for i in range(5):
    print('Это', i, 'раз когда мы выполнили этот блок кода')
````

Цикл `while`:

````python
i = 0
while i < 5:
    print('Это', i, 'раз когда мы выполнили этот блок кода')
    i = i + 1 # можно написать i += 1
````

В `range` можем указывать начало, конец и шаг:
``` python
for *любое имя переменной* range(старт, стоп, шаг): 
    *код который будет повторяться*
```
Можно делать цикл в цикле (*вложенный цикл*):
``` python
for *любое имя переменной* in range(*положительное целое число*): 
    *код который будет повторяться*
    for *любое (другое) имя переменной* in range(*положительное целое число*): 
        *код который будет повторяться вообще много раз*
*код вне цикла*
```

### Ключевые слова в циклах

Можно использовать следующие ключевые слова для большего контроля:
- `continue` используется если нужно пропустить итерацию цикла
- `break` используется если нужно выйти из цикла
- `pass` используется когда ничего делать не надо

# **Списки**

*Списки* (`list` или *листы*) это особый тип данных: в нем может храниться сразу несколько переменных.

Пример списка:

````python
checklist = ['Стаканчики', 'Мандарины', 'Хлеб', 'Колбаса', 'Сыр'] 
````

Синтаксис: 
``` python
*список* = [*элемент*, *элемент*, *элемент*]
```

Каждый элемент имеет свой *индекс*, он начинается с нуля и позволяет выбрать нужный нам элемент списка. Обращение по индексу заключается в квадратных скобочках:

````python
for i in range(5):
    print(str(i) + ' элемент списка: ' + checklist[i])
````

Для индексирования можно использовать *срезы*:

````python
print('Элементы от нулевого (включительно) до конечного (невключительно): ', checklist[0:2])  
print('То же самое, если считаем от нуля то можно опустить первую цифру в скобках: ', checklist[:2]) 
print('Элементы от второго до последнего (включительно): ', checklist[2:]) 
print('Элементы от второго до последнего (невключительно): ', checklist[2:-1])
print('-1 это последний элемент списка: ', checklist[-1]) # строго говоря это не срез, а просто элемент
print('Можно считать с конца, -2 это предпоследний: ', checklist[-2]) # строго говоря это не срез, а просто элемент
print('Можно индексировать с шагом: ', checklist[0::2])
````

Можно и так использовать циклы:

````python
for element in checklist:
    print('Элемент списка: ' + element)
````
